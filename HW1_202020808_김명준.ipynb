* data에서 Education 열의 데이터 값을 `Graduation`일 경우 0, `PhD`일 경우 1, `Master`일 경우 2, `Basic`일 경우 3, `2n Cycle`일 경우 4로 인코딩하세요.

```python
################  문제 2-1  ####################
# 소스코드를 입력하고 캡처 후 보고서에 업로드하세요.
# 코드 시작과 코드 끝 라인을 삭제하지 마세요.
################ 코드 시작  #####################

# Gender 열의 unique()의 결과에서 순서대로 0부터 값을 부여
data['Gender'] = data['Gender'].factorize()[0]
# Customer Type 열의 원소 중 VIP는 1로 common은 0으로 매핑
data['Customer Type'] = data['Customer Type'].map({"VIP":1, "common":0})
# Marital_Status 열의 unique()의 결과에서 순서대로 0부터 값을 부여
data['Marital_Status'] = data['Marital_Status'].factorize()[0]
# Education 열의 unique()의 결과에서 순서대로 0부터 값을 부여
data['Education'] = data['Education'].factorize()[0]

##################################################
```

### 문제 2-2
오늘 날짜는 2024년 3월 16일 입니다.

`Year_Birth` 열은 현재 **년도** 단위로 데이터가 구성되어 있습니다.

**나이** 단위로 데이터를 변경하고 캡처하여 보고서에 업로드 하세요.

예)  1957년도 출생 => 2024-1957=67

`DT_Customer` 열은 현재 **날짜** 단위로 데이터가 구성되어 있습니다.  

**년** 단위로 데이터를 변경하고 캡처하여 보고서에 업로드 하세요.

예) 04-09-2012 => (03-16-2024-04-09-2012)/365=11.94 (소수점 셋째자리에서 버림)

**Hint) datetime.strptime**

```python
from datetime import datetime
################  문제 2-2  ####################
# 소스코드를 입력하고 캡처 후 보고서에 업로드하세요.
# 코드 시작과 코드 끝 라인을 삭제하지 마세요.
################ 코드 시작  #####################

# Year_Birth 열의 값을 2024 - 현재값으로 변경
data['Year_Birth'] = 2024 - data['Year_Birth']
# Dt_Customer 열의 타입을 string으로 변경
data['Dt_Customer'] = data['Dt_Customer'].astype(str)
# Dt_Customer 열의 타입을 formating을 통해 strptime 변경 후 그 차이의 days를 저장하여 일 수를 저장
data['Dt_Customer'] = data['Dt_Customer'].apply(lambda _ :
 (datetime.strptime('03-16-2024', '%m-%d-%Y') - datetime.strptime(_,'%d-%m-%Y')).days)
# 365일로 값을 나누고 100을 곱한 후 버림한 것을 다시 100을 곱해 소수점 둘째자리까지 저장
data['Dt_Customer'] = np.floor((data['Dt_Customer'] / 365) * 100) / 100

################ 코드 끝 ########################
#################################################
```

### 문제 3-2
주어진 코드틀에 k개의 "MntToyProducts" 값을 보고 평균을 내어 결측값을 결정하는 코드를 작성하세요.
* 주의) 코드틀 이외의 부분에 코드를 작성하지 마세요!
* 주의) 소스코드를 입력하고 캡처 후 보고서에 업로드하세요.
* 주의) 코드 시작과 코드 끝 라인을 삭제하지 마세요.

```python
from tqdm.notebook import tqdm
```

```python
class Knn_imputer():
  def __init__(self, n_neighbors):

    ## 생성자로 최근접 이웃수(n_neighbors)를 입력 받습니다.
    self.k = n_neighbors

  ## knn 은 lazy learning 입니다.
  def fit(self,X):
      self.X = X

  def predict(self, X):

    for i in tqdm(range(len(X))):
     if pd.isnull(X['MntToyProducts'].iloc[i]): # Arrival Delay 열에 있는 결측값 확인

        distance = self.dist(X, X.iloc[i]) # 데이터프레임 X와 결측값이 있는 행 데이터 간의 거리 구하는 함수
        sorted_dist = sorted(distance.items(), key=lambda x: x[1]) # distance 를 sorting 하기

         ################  문제 3-2  ####################
         # k개의 "MntToyProducts" 값을 보고 평균 내어
         # 값을 결정하기

         # 소스코드를 입력하고 캡처 후 보고서에 업로드하세요.
         # 코드 시작과 코드 끝 라인을 삭제하지 마세요.
        ################ 코드 시작  #####################

        # 거리의 합을 저장하는 변수와 저장 횟수를 나타내는 변수
        sum ,cnt = 0, 0
        # sorted_dist를 형변환 후 순차적으로, 가까운 것부터 탐색
        for key in dict(sorted_dist).keys():
          # 자신의 값을 제외한 가까운 것부터 MntToyProducts 열의 값을 더함
          if key != i :
            sum = sum + X['MntToyProducts'][key]
            cnt = cnt + 1
          # 횟수가 k가 되면 중지
          if cnt == self.k :
            break

        # sum을 k로 나누어 평균을 구하기
        X['MntToyProducts'][i] = sum / self.k

        ################ 코드 끝  ######################
        ##############################################

    return X['MntToyProducts']

  def dist(self, X, y):
    num_train = X.shape[0]

    X = X.drop(['MntToyProducts'], axis=1)
    y = y.drop(['MntToyProducts'])
    dist_dict = {} ## {index : distance} 형태

    for i in range(num_train):

    ################  문제 3-1  ####################
    # 데이터프레임 X와 결측값이 있는 행 데이터 간의 거리 구하기
    # 소스코드를 입력하고 캡처 후 보고서에 업로드하세요.
    # 코드 시작과 코드 끝 라인을 삭제하지 마세요.
    ################ 코드 시작  #####################

    # 인덱스를 의미하는 dist_dict의 키 i에 MntToyProducts 이외의 열 간 거리를 저장
    # 즉 Kidhome 열의 거리^2을 저장
      dist_dict[i] = (X['Kidhome'][i] - y['Kidhome'])**2

    ################ 코드 끝  ######################
    ##############################################

    return dist_dict
```

```python
imputer = Knn_imputer(3)
imputer.fit(X)

# test set에 대해 predict를 진행합니다.
data['MntToyProducts'] = imputer.predict(X) # 오래 걸립니다
```

### 문제 5-1
**`KNeighborsClassifier`**를 사용하여 주어진 조건에 맞게 모델을 학습하고 완성한 코드를 보고서에 캡처하여 업로드하세요.
* 조건 1: neighbor의 수 `k= [1, 3, 5, 15, 55, 101, 501]`으로 변화에 따른 정확도의 변화를
`line plot`으로 그리고 가장 높은 성능의 `k`를 출력하시오.
* 조건 2: train set에서 validation set을 나누세요. `train : validation = 7 : 3`

* *주의! : test set이 아닌 validation set으로 평가해야 합니다.*

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
################  문제 5-1  ####################
# 소스코드를 입력하고 캡처 후 보고서에 업로드하세요.
# 코드 시작과 코드 끝 라인을 삭제하지 마세요.
################ 코드 시작  #####################

# neighbor의 수를 의미하는 리스트
numK = [1, 3, 5, 15, 55, 101, 501]
# KNeighborsClassifier의 결과 점수를 저장하는 리스트
knn_scores = []
# X_train을 다시 7대3으로 분할하여 validation set 만들기
X_train2, X_valid, y_train2, y_valid = train_test_split(X_train, y_train, test_size=0.3)

# neighbor의 수마다 체크
for e in numK:
  # KNeighborsClassifier를 통해 분류기 설정
  classifier = KNeighborsClassifier(n_neighbors = e)
  # 트레이닝 셋을 통해 트레이닝
  classifier.fit(X_train2, y_train2)
  # 벨리데이션 셋에 대한 분류기의 점수를 저장
  knn_scores.append(classifier.score(X_valid, y_valid))

# knn 스코어의 최대값을 가지는 neighbor의 수 출력
print(numK[knn_scores.index(max(knn_scores))])

################ 코드 끝  ######################
##############################################
```

```python
sns.lineplot(x = numK, y = knn_scores)
```

### 문제 5-2
**`DecisionTreeClassifier`**를 사용하여 주어진 조건에 맞게 모델을 학습하고 완성한 코드 및 출력물을 보고서에 캡처하여 업로드하세요.
* 조건 1: 최대 트리의 깊이를 `depth  = [1, 3, 5, 10, 15, 30, 20]`으로 변화에 따른 정확도의 변화를 `line plot`으로 그리고 가장 높은 성능의 `depth`를 출력하시오.
* 조건 2: train set에서 validation set을 나누세요. `train : validation = 7 : 3`

* *주의! : test set이 아닌 validation set으로 평가해야 합니다.*

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
################  문제 5-2  ####################
# 소스코드를 입력하고 캡처 후 보고서에 업로드하세요.
# 코드 시작과 코드 끝 라인을 삭제하지 마세요.
################ 코드 시작  #####################

# 최대 트리의 깊이를 의미하는 리스트
num_dep = [1, 3, 5, 10, 15, 30, 20]
# DecisionTreeClassifier의 점수를 저장하는 리스트
dt_scores = []
# X_train을 다시 7대3으로 분할하여 validation set 만들기
X_train2, X_valid, y_train2, y_valid = train_test_split(X_train, y_train, test_size=0.3)

# 최대 트리의 깊이를 변화시키면서 체크
for e in num_dep:
  # DecisionTreeClassifier를 통해 분류기 설정
  classifier = DecisionTreeClassifier(max_depth = e)
  # 트레이닝 셋을 통해 트레이닝
  classifier.fit(X_train2, y_train2)
  # 벨리데이션 셋에 대한 분류기의 점수를 저장
  dt_scores.append(classifier.score(X_valid, y_valid))

# DecisionTree 스코어의 최대값을 가지는 최대 트리의 깊이 출력
print(num_dep[dt_scores.index(max(dt_scores))])

################ 코드 끝  ######################
##############################################
```

```python
sns.lineplot(x = num_dep, y = dt_scores)
```

### 문제 5-3
`문제 5-1` 과 `문제 5-2`에서 찾은 best 하이퍼파라미터 (k, max depth)로 두 분류기를 전체 train set 다시 각각 학습하여 test set으로 평가하고 완성한 코드와 출력을 보고서에 캡처하여 업로드하세요.

* 위에서 찾은 최적의 하이퍼파라미터 선택
* 전체 train set으로 분류기 학습
* Test set을 통해서 평가 및 비교

```python
################  문제 5-2  ####################
# 소스코드를 입력하고 캡처 후 보고서에 업로드하세요.
# 코드 시작과 코드 끝 라인을 삭제하지 마세요.
################ 코드 시작  #####################

# knn 스코어의 최대값을 가지는 neighbor의 수
best_k = numK[knn_scores.index(max(knn_scores))]
# DecisionTree 스코어의 최대값을 가지는 최대 트리의 깊이
best_depth = num_dep[dt_scores.index(max(dt_scores))]

# 최대값을 가지는 neighbor의 수의 분류기로 전체 트레이닝 셋 학습
Knn_classifier = KNeighborsClassifier(n_neighbors = best_k)
Knn_classifier.fit(X_train, y_train)

# 최대값을 가지는 최대 트리 깊이의 분류기로 전체 트레이닝 셋 학습
DT_classifier = DecisionTreeClassifier(max_depth = best_depth)
DT_classifier.fit(X_train, y_train)

# 테스트 셋에 대한 점수를 출력
Knn_acc = Knn_classifier.score(X_test, y_test)
DT_acc = DT_classifier.score(X_test, y_test)
print(Knn_acc)
print(DT_acc)
################ 코드 끝  ######################
##############################################
```
